```python
#import libraries
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
#Read in the data

df =pd.read_csv('https://raw.githubusercontent.com/danielgrijalva/movie-stats/master/movies.csv')
```

```python
#Look at the data header

df.head()
```

```python
#If there are any missing data

for col in df.columns:
    pct_missing=np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Let's drop rows that having missing values

df = df.dropna()
```

```python
#Let's check again if there are any missing values

for col in df.columns:
    pct_missing=np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data type for our columns

df.dtypes
```

```python
#change data type of columns 'votes,budget,gross'

df['votes'] = df['votes'].astype('int64')
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
#sort data by revenue 'gross'

pd.set_option('display.max_rows', None)

df.sort_values(by=['gross'], inplace =False, ascending =False).head()
```

## Let's put a hypothesis that there are a correlation between the budget anf the gross revenue.

```python
# Scatter plot budget vs gross revenue

plt.scatter(x=df['budget'],y=df['gross'])

plt.title('Budget vs Gross revenue')

plt.xlabel('Gross revenue')

plt.ylabel('Budget of the movie')

plt.show()
```

```python
#Plot budget vs gross using seaborn

sns.regplot(x='budget',y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# Looking for correlation using pearson method 

df.corr(method ='pearson')
```

# Conclusion: There is a High correlation between budget and gross

```python
correlation_matrix = df.corr(method ='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matric for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
    
df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(method ='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matric for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_matrix_num = df_numerized.corr()

corr_pairs = correlation_matrix_num.unstack()

sorted_pairs = corr_pairs.sort_values()

high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr



```
